```python
stop_times['arrival_hour'] = stop_times['arrival_time'].astype(int) // 1800
trips_per_hhour = stop_times.groupby('arrival_hour').size()

plt.figure(figsize=(12, 6))
plt.plot(trips_per_hhour.index, trips_per_hhour.values, marker='o', linestyle='-')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Stops')
plt.title('Peak 30-min Analysis (Based on Seconds Since Midnight)')
plt.xticks(range(0, 49, 4))
plt.grid(True)
plt.show()
```

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371
    return c * r

stops_coordinates = {
    'Stop ID': ['S1', 'S2', 'S3', 'S4'],
    'Latitude': [40.748817, 40.748441, 40.748939, 40.749641],
    'Longitude': [-73.985428, -73.985664, -73.984540, -73.983776]
}

stops_df = pd.DataFrame(stops_coordinates)

distances_km = []
for i in range(len(stops)-1):
    lat1, lon1 = stops.iloc[i]['stop_lat'], stops.iloc[i]['stop_lon']
    lat2, lon2 = stops.iloc[i+1]['stop_lat'], stops.iloc[i+1]['stop_lon']
    distance = haversine(lon1, lat1, lon2, lat2)
    distances_km.append(distance)

stops['Distance to Next Stop (km)'] = distances_km + [None]  # Adding None for the last stop without a next stop
```

```python
n_bins_lat = 60
n_bins_lon = int(n_bins_lat * (stops['stop_lon'].min() - stops['stop_lon'].max()) / (stops['stop_lat'].min() - stops['stop_lat'].max()) )
lat_bins = np.linspace(stops['stop_lat'].min(), stops['stop_lat'].max(), n_bins_lat+1)
lon_bins = np.linspace(stops['stop_lon'].min(), stops['stop_lon'].max(), n_bins_lon+1)

hist, lat_edges, lon_edges = np.histogram2d(stops['stop_lat'], stops['stop_lon'], bins=[lat_bins, lon_bins])

fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.imshow(np.log1p(hist), interpolation='nearest', origin='lower',
                extent=[lon_edges[0], lon_edges[-1], lat_edges[0], lat_edges[-1]],
                aspect='auto')

cbar = fig.colorbar(cax, ax=ax, orientation='vertical')
cbar.set_label('Log of Number of Trips + 1')

ax.set_title('Number of Trips by Geographic Area')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px
dir_0 = speeds.loc[(speeds.direction_id==0)&(speeds.route_id=='1')].sort_values(by='stop_sequence')
dir_0['hour'] = dir_0.window.apply(lambda x: int(x.split(':')[0]))
dir_0.sort_values(by='hour', ascending=True, inplace=True)

fig = go.Figure(data=go.Heatmap(
                   z=dir_0.speed_kmh,
                   y=dir_0.start_stop_name,
                   x=dir_0.window,
                   hoverongaps = False,
                   colorscale=px.colors.colorbrewer.RdYlBu, 
                   reversescale=False
))

fig.update_yaxes(title_text='Stop', autorange='reversed')
fig.update_xaxes(title_text='Hour of day', side='top')
fig.update_layout(showlegend=False, height=600, width=1000,
                 title='Speed heatmap per direction and hour of the day')

fig.show()
```
